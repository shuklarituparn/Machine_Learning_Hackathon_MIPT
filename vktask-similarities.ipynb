```python
splits = {'train': 'data/train-00000-of-00001-d1de06d6b6482a7a.parquet', 'dev': 'data/dev-00000-of-00001-6014228512208659.parquet', 'testA': 'data/testA-00000-of-00001-78a4a011ae1dc086.parquet', 'testB': 'data/testB-00000-of-00001-003543ac41b77cc0.parquet'}
df = pd.read_parquet("hf://datasets/Cohere/miracl-ru-queries-22-12/" + splits["train"])
```

```python
from datasets import load_dataset
ds = load_dataset("Cohere/miracl-ru-queries-22-12")
```

```python
train = ds["train"]
dev = ds["dev"]
train_queries = train["query"]
val_queries = dev["query"]
```

```python
similar_sentences = 0
for train in train_queries:
    for val in val_queries:
        if val == train:
            similar_sentences +=1
print("Количество одинаковых вопросов в тренировочном датасете и валидационном датасете равна ", similar_sentences)
```
